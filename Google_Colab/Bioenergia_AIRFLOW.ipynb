###Estipula que a Coluna Mês/Ano seja do Tipo String

```python
df['Mês/Ano'] = df['Mês/Ano'].astype(str)
```

###Tratamento da Coluna Mês/Ano, retornando apenas os dados significativos

```python
df["Mês/Ano"] = df["Mês/Ano"].str.slice(0 ,7)
```

###Exclui a Coluna Região

```python
# A coluna Região foi excluída, pois na tabela existem outras colunas que ajudarão na geolocalização dos endereços

df = df.drop(['Região'], axis=1)
```

###Lista Todos os Valores Únicos da Coluna Mês/Ano

```python
# Utilização da função Sorted para listar todos os valores da coluna Mês/Ano
# Função sorted, ordena as informações da lista apresentado os resultados em 20 colunas

data =  sorted(pd.unique(df['Mês/Ano']))
for d in range(len(data)):
  if d % 20 == 0:
        print()
  print(data[d], end=', ')

```

###Estipula que a Coluna Razão Social seja do Tipo String

###Lista Todos os Valores Únicos da Coluna Razão Social

```python
# Utilização da função Sorted para listar todos os valores da coluna Razão Social
# Função sorted, ordena as informações da lista apresentado os resultados em 1 coluna

razao =  sorted(pd.unique(df['Razão Social']))
for r in range(len(razao)):
  if r % 1 == 0:
        print()
  print(razao[r], end=', ')

  # inconsistência nan
```

###Estipula que a Coluna Produto seja do Tipo String

```python
df['Produto'] = df['Produto'].astype(str)
```

###Lista Todos os Valores Únicos da Coluna Produto

```python
# Utilização da função Sorted para listar todos os valores da coluna Produto
# Função sorted, ordena as informações da lista apresentado os resultados em 1 coluna

produto =  sorted(pd.unique(df['Produto']))
for p in range(len(produto)):
  if p % 1 == 0:
        print()
  print(produto[p], end=', ')

  # inconsistências de nan
```

###Replace de Inconsistência na Coluna Produto

```python
df['Produto'] = df['Produto'].apply(lambda x: x.replace('(ELAEIS GUINEENSIS OU ELAEIS O', '(ELAEIS GUINEENSIS OU ELAEIS O)'))
```

###Estipula que a Coluna Quantidade (m³) seja do Tipo Float

```python
df['Quantidade (m³)'] = df['Quantidade (m³)'].astype(float)
```

###Lista Todos os Valores Únicos na Coluna Quantidade (m³)

```python
# Utilização da função Sorted para listar todos os valores da coluna Quantidade (m³)
# Função sorted, ordena as informações da lista apresentado os resultados em 1 coluna

qtde =  sorted(pd.unique(df['Quantidade (m³)']))
for q in range(len(qtde)):
  if q % 1 == 0:
        print()
  print(qtde[q], end=', ')
```

###Tratamento de Inconsistências Encontradas nas Colunas

```python
# Tratamento de Inconsistências

df.replace(["NaN", "nan", " ", "", "NAN", "NA"], pd.NA, inplace = True)
```

### Groupby que retorna a Quantidade total dos Produtos

```python
dfprod = df.groupby(['Estado', 'Produto'])['Quantidade (m³)'].sum().reset_index()
```
